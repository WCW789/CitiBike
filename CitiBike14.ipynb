```python
import pandas as pd
import numpy as np
```

```python
file_05_14 = "201405-CitiBike_tripdata.xlsx"
```

```python
file_05_14_df = pd.read_excel(file_05_14, index_col=None)
file_05_14_df.head()
```

```python
start_data = pd.DatetimeIndex(file_05_14_df['starttime'])
print(start_data)

end_data = pd.DatetimeIndex(file_05_14_df['stoptime'])
print(end_data)
```

```python
file_05_14_df['Start_Date'] = start_data.date
file_05_14_df['Start_Date'].head(2)
```

```python
file_05_14_df['Start_Time'] = start_data.time
file_05_14_df['Start_Time'].head(2)
```

```python
file_05_14_df['End_Date'] = end_data.date
file_05_14_df['End_Date'].head(2)
```

```python
file_05_14_df['End_Time'] = end_data.time
file_05_14_df['End_Time'].head(2)
```
