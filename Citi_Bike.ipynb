```python
import pandas as pd
import numpy as np
```

```python
file_05 = "201905-citibike-tripdata.xlsx"
```

```python
file_05_df = pd.read_excel(file_05, index_col=None)
file_05_df.head()
```

```python
start_data = pd.DatetimeIndex(file_05_df['starttime'])
print(start_data)

end_data = pd.DatetimeIndex(file_05_df['stoptime'])
print(end_data)
```

```python
file_05_df['Start_Date'] = start_data.date
file_05_df['Start_Date'].head(2)
```

```python
file_05_df['Start_Time'] = start_data.time
file_05_df['Start_Time'].head(2)
```

```python
file_05_df['End_Date'] = end_data.date
file_05_df['End_Date'].head(2)
```

```python
file_05_df['End_Time'] = end_data.time
file_05_df['End_Time'].head(2)
```

```python
file_05_df[["Start_Date", "Start_Time", "End_Date", "End_Time"]].head(2)
```
